## Gráficas de densidad

- Empleadas para representar el valor asociado a un par de variables (x,y)
- Como ejemplo, vamos a gráficar los datos contenidos en el archivo data/circular.txt. Estos datos corresponden a una matriz de 501x501 valores, donde cada elemento representa el valor de la posición (i,j)

```python
from pylab import imshow, show, xlabel, ylabel
from numpy import loadtxt

data = loadtxt("../data/circular.txt", float)
print(data.shape)
imshow(data)
show()

print("hola")
```

### Note que los indices son contrarios a la convención (x, y). Es decir, primero se barren las columnas, se empieza con el índice i, y luego se barren las filas, índice j.

- En python se puede especificar lo contrario: imshow(data, origin="lower")

```python
imshow(data, origin="lower")
xlabel("Index i")
ylabel("Index j")
show()
```

### Tambien se puede hacer en escala de grisis

```python
from pylab import gray, xlim, ylim
imshow(data, origin="lower")
gray()
#xlim(0., 50.)
#ylim(0., 50.)
show()
```

```python
from pylab import hot
imshow(data, origin="lower")
hot()
show()
```

### También se puede cambiar y/o calibrar la escala

```python
imshow(data, origin="lower", extent=[0, 10, 0, 5])
show()
```

- También podemos decir a imshow cómo es la relación de aspecto, esto es, cuánto es la proporción del eje verical respecto del horizontal

```python
imshow(data, origin="lower", extent=[0, 10, 0, 5], aspect=2.0)
show()
```
