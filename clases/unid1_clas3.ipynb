## Gráficos: 
- Vamos usar Pylab, un paquete que hace parte de uno más grande llamado matplotlib.
- Dos tipos de gráficos: lineales, dispersión, y gráficos de densidad.

```python
from pylab import plot, show

y = [1., 2.4, 1.7, 0.3, 0.6, 1.8]

plot(y)
show()
```

En este caso Python asignó la escala del eje X.
<br>
Para especificar esta escala:

```python
x = [0.5, 1., 2., 4., 7., 10.]

plot(x,y)
show()
```

Con la opción plot() podemos insertar varios plot en el mismo gráfico.
<br><br>
Para salvar el gráfico en un archivo:

```python
from pylab import savefig
plot(x, y)
savefig('destination_path.eps', format='eps', dpi=1000)
show()
```

Para graficar una función conocida, por ejemplo la función Sin:

```python
from numpy import linspace, sin, cos

x = linspace(0, 10, 1000)
y = sin(x)

plot(x,y)
show()
```

Note que en este caso, la función sin actuó sobre cada elemento del array x.
De manera equivalente a map(sin, x)

## La escala de los ejes se puede modificar

```python
from pylab import ylim, xlim, grid
x = linspace(0, 10, 1000)
y = sin(x)

plot(x,y)
ylim(-1.1, 1.1)
xlim(0., 10)
grid("on")
show()
```

- También podemos insertar las etiquetas de los ejes

```python
from pylab import xlabel, ylabel
plot(x,y)
ylim(-1.1, 1.1)
xlim(0., 10)
grid("on")
xlabel("Eje X")
ylabel("Eje Y")
show()
```

## Diferentes formas de gráficar

```python
plot(x,y,'g--')
show()
```

```python
x = linspace(0, 10, 50)
y = sin(x)


plot(x,y,'ko')
show()
```

```python
y = sin(x)
plot(x,y,'k-')
y = cos(x)
plot(x,y,'k--')
grid("on")
show()
```
