## Dos maneras de hacerlo
- plot
- scatter

```python
from pylab import plot, scatter, xlabel, ylabel, xlim, ylim, show
from numpy import loadtxt

data = loadtxt("../data/stars.txt", float)
plot(data[:,0], data[:,1], 'ko')
show()
```

```python
scatter(data[:,0], data[:,1])
show()
```

```python
scatter(data[:,0], data[:,1])

xlabel("Temperature")
ylabel("Magnitude")
xlim(13000,0)
ylim(20,-5)
show()
```
